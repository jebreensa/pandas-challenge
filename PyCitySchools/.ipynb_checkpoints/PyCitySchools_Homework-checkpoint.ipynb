```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Total Number of Schools
School_Names_df=school_data["school_name"].unique()
Total_Num_Schools= len(School_Names_df)
#Total Number of Students
Student_Name_df=student_data["Student ID"].unique()
Total_Num_Students= len(Student_Name_df)
#Total Budget
Total_Budget=school_data["budget"].sum()
#Average Math Score
Average_Math_Score=student_data["math_score"].mean()
#Average Reading Score
Average_Reading_Score=student_data["reading_score"].mean()
#Percentage of students with a passing math score(70 or greater)
Percentage_Math_Passing_Score= ((student_data["math_score"]>=70)/Total_Num_Students).sum()
#Percentage of students with a passing reading score(70 or greater)
Percentage_Reading_Passing_Score= ((student_data["reading_score"]>=70)/Total_Num_Students).sum()
#Percentage of Overall Passing
Percentage_Overall_Passing= (((student_data["math_score"]>=70)& (student_data["reading_score"]>=70))/Total_Num_Students).sum()
#Summary Results
Summary_Results= pd.DataFrame({"Total_Num_Schools":[Total_Num_Schools],
                              "Total_Num_Students":[Total_Num_Students],
                              "Total_Budget":[Total_Budget],
                              "Average_Math_Score":[Average_Math_Score],
                              "Average_Reading_Score":[Average_Reading_Score],
                              "Percentage_Math_Passing_Score":[Percentage_Math_Passing_Score],
                              "Percentage_Reading_Passing_Score":[Percentage_Reading_Passing_Score],
                               "Percentage_Overall_Passing":[Percentage_Overall_Passing]})
Summary_Results.style.format({"Total_Budget":"${:,.2f}",
                             "Average_Math_Score": "{:.1f}",
                             "Average_Reading_Score": "{:.1f}",
                             "Percentage_Math_Passing_Score": "{:.1%}",
                             "Percentage_Reading_Passing_Score": "{:.1%}",
                             "Percentage_Overall_Passing": "{:.1%}"})
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#group by School
Group_By_School=school_data_complete.set_index("school_name").groupby(["school_name"])
#SchoolType
School_Type =school_data.set_index("school_name")["type"]
#Total Students
Students_Per_School= Group_By_School["Student ID"].count()
#Total Budget
Total_Budget_Per_School=school_data.set_index("school_name")["budget"]
#Per Student Budget
Per_Student_Budget=Total_Budget_Per_School/Students_Per_School
#Average Math Score
Average_Math_Score_Per_School=Group_By_School["math_score"].mean()
#Average Reading Score
Average_Reading_Score_Per_School=Group_By_School["reading_score"].mean()
#% passing Math
Passing_Math_Per_School= school_data_complete[school_data_complete["math_score"]>=70].groupby("school_name")["Student ID"].count()/Students_Per_School
Passing_Reading_Per_School= school_data_complete[school_data_complete["reading_score"]>=70].groupby("school_name")["Student ID"].count()/Students_Per_School
Passing_OverAll_Per_School= school_data_complete[(school_data_complete["math_score"] >=70) & (school_data_complete["reading_score"]>=70)].groupby("school_name")["Student ID"].count()/Students_Per_School

#School Summary 

School_Summary= pd.DataFrame({"School_Type":School_Type,
                              "Students_Per_School":Students_Per_School,
                              "Total_Budget_Per_School":Total_Budget_Per_School,
                              "Average_Math_Score_Per_School":Average_Math_Score_Per_School,
                              "Average_Reading_Score_Per_School":Average_Reading_Score_Per_School,
                              "Passing_Math_Per_School":Passing_Math_Per_School,
                              "Passing_Reading_Per_School":Passing_Reading_Per_School,
                               "Passing_OverAll_Per_School":Passing_OverAll_Per_School})

School_Summary.style.format({"Total_Budget_Per_School":"${:,.0f}",
                             "Average_Math_Score_Per_School": "{:.1f}",
                             "Average_Reading_Score_Per_School": "{:.1f}",
                             "Passing_Math_Per_School": "{:.1%}",
                             "Passing_Reading_Per_School": "{:.1%}",
                             "Passing_OverAll_Per_School": "{:.1%}"})
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
#Top Performing Schools
Top_Performing_School=School_Summary.sort_values(["Passing_OverAll_Per_School"],ascending=False)
Top_Performing_School=Top_Performing_School.head().style.format({"Total_Budget_Per_School":"${:,.0f}",
                             "Average_Math_Score_Per_School": "{:.1f}",
                             "Average_Reading_Score_Per_School": "{:.1f}",
                             "Passing_Math_Per_School": "{:.1%}",
                             "Passing_Reading_Per_School": "{:.1%}",
                             "Passing_OverAll_Per_School": "{:.1%}"})
Top_Performing_School
```

## Bottom Performing Schools (By % Overall Passing)

* Sort and display the five worst-performing schools by % overall passing.

```python
#Bottom Performing Schools
Bottom_Performing_School=School_Summary.sort_values(["Passing_OverAll_Per_School"],ascending=True)
Bottom_Performing_School=Bottom_Performing_School.head().style.format({"Total_Budget_Per_School":"${:,.0f}",
                             "Average_Math_Score_Per_School": "{:.1f}",
                             "Average_Reading_Score_Per_School": "{:.1f}",
                             "Passing_Math_Per_School": "{:.1%}",
                             "Passing_Reading_Per_School": "{:.1%}",
                             "Passing_OverAll_Per_School": "{:.1%}"})
Bottom_Performing_School
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Scores By School Spending 
Bins_Range=[0,600,620,650,10000]
Binned_Group=["0-600","600-620","620-650","650-10000"]
school_data_complete["Average Spending Ranges"]=pd.cut(school_data_complete["budget"]/school_data_complete["size"],Bins_Range,labels=Binned_Group)

By_School_sepnding= school_data_complete.groupby("Average Spending Ranges")
Average_Math= By_School_sepnding["math_score"].mean()
Average_Reading= By_School_sepnding["reading_score"].mean()
Passing_Math= school_data_complete[school_data_complete["math_score"]>=70].groupby("Average Spending Ranges")["Student ID"].count()/By_School_sepnding["Student ID"].count()
Passing_Reading= school_data_complete[school_data_complete["reading_score"]>=70].groupby("Average Spending Ranges")["Student ID"].count()/By_School_sepnding["Student ID"].count()
Passing_OverAll= school_data_complete[(school_data_complete["math_score"] >=70) & (school_data_complete["reading_score"]>=70)].groupby("Average Spending Ranges")["Student ID"].count()/By_School_sepnding["Student ID"].count()

Scores_By_Spending= pd.DataFrame({"Average_Math":Average_Math,
                                  "Average_Reading":Average_Reading,
                                  "Passing_Math":Passing_Math,
                                  "Passing_Reading":Passing_Reading,
                                 "Passing_OverAll":Passing_OverAll})

Scores_By_Spending=Scores_By_Spending.style.format({"Average_Math":"{:.1f}",
                                  "Average_Reading":"{:.1f}",
                                  "Passing_Math":"{:.1%}",
                                  "Passing_Reading":"{:.1%}",
                                 "Passing_OverAll":"{:.1%}"})
Scores_By_Spending
```
